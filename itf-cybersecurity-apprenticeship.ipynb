###### Motherboard Components
***

***
###### CPU
***

The ``Central Processing Unit`` handles all instructions it receives from hardware and software running on the computer.

***
###### RAM
***

``Random Access Memory`` or RAM is the physical hardware inside a computer that temporarily stores data, serving as the computer's working memory. 

***
###### Northbridge
***

``Northbridge`` or PAC (PCI/AGP Controller) is an integrated circuit responsible for communications between the CPU interface, Accelerated Graphics Port (AGP), and the memory. Unlike the ``Southbridge``, the ``Northbridge`` directly connects to these components, acting as a ``bridge`` for the Southbridge Chip to communicate with the CPU, RAM, and graphics controller.

***
###### Southbridge
***

The ``Southbridge`` is an integrated circuit on a motherboard responsible for the hard drive controller, I/O controller and integrated hardware (including the sound card and video card if on the motherboard, USB, PCI, ISA, IDE, BIOS, and Ethernet). 

It gets its name for commonly being South of the PCI bus.

***
###### Network interface Card - NIC
***

A ``Network Interface Card`` is a circuit board or chip installed on a computer so it can connect to a network - acts as the interface between machine and network.

***
###### PCI Slot
***

``Peripheral Component Interconnect``  are standard connection interfaces for attaching computer peripherals to the motherboard - plugging specialized parts within computer devices.

***
###### Heat Sink
***

Incorporates a fan or another mechanism to reduce the temperature of a hardware component (e.g., processor). 

***
###### Land Grid Array - ``mainly used by Intel``
***

``Land Grid Array`` (``LGA``) is an integrated circuit design involving a square grid of contacts connected to other components of a printed circuit board.

***
###### Pin Grid Array - ``mainly used by AMD now``
***

The contact pins on a ``Pin Grid Array``are on the CPU, while the socket on the motherboard contains holes to secure the CPU.

***
###### Binary - Bit to TB
***

Binary is the most straightforward number system possible - only two digits: 0 and 1 (off / on). A bit represents a binary digit and is the basic building block of information. Given a bit is the tiniest amount of information possible, anything less than a bit is no information. Using multiple bits can convey more complex information.

> 8 bits = 1 Byte

> 1,024 Bytes = 1 KB

> 1,024 KB = 1 MB

> 1,024 MB = 1 GB

> 1,024 GB = 1 TB
